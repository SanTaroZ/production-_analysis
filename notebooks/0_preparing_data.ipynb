# CLEANING DATA

```python
#Filter only useful columns
data = data [['FECHA', 'TURNO', 'GRUPO', 'MÁQ.', 'CODIGO', 'PRODUCCION (PZAS)',
       'TOTAL KG', 'SCRAP']]
data
```

```python
type(data)
```

```python
data.dtypes
```


 *   Cleaning Column "FECHA"
 

```python
data = data[data["FECHA"].notna()] #cleaning NA
data
```

```python
data = data[data["FECHA"] != "Total"] #Cleaning Row "Total"
data
```

```python
data['FECHA'] = pd.to_datetime(data['FECHA'], format= '%Y-%m-%d')
data
```

```python
data.dtypes
```

```python
data['FECHA'].dt.day # day
```

```python
data['FECHA'].dt.day.unique() 
```

```python
data['DIA'] = data.loc[:,'FECHA'].dt.day
data
```

```python
data['NOMBRE_DIA']=data.loc[:,'FECHA'].dt.day_name()
data
```

```python
data.dtypes
```

* Setting column TURNO to category

```python
data['TURNO'] = data['TURNO'].astype('category')
data.dtypes
```

* Setting column GRUPO to Category

```python
data['GRUPO'] = data['GRUPO'].astype('category')
data.dtypes
```

```python
data['GRUPO'].unique()
```

```python
data
```

* Cleaning column CODIGO

```python
data.loc[:,'CODIGO'] = data['CODIGO'].fillna(0)
```

```python
data
```

* Cleaning column PRODUCCION (PZAS)

```python
data[data['PRODUCCION (PZAS)'].isnull()]
```

```python
data['PRODUCCION (PZAS)'] = data['PRODUCCION (PZAS)'].fillna(0)
data
```

```python
data[data['PRODUCCION (PZAS)'].isnull()]
```

```python
data['PRODUCCION (PZAS)'] = data['PRODUCCION (PZAS)'].astype('int64')
data.dtypes
```

* Cleaning column TOTAL KG

```python
data.dtypes
```

```python
data.loc[:,'TOTAL KG'] = data['TOTAL KG'].astype('float')
data.dtypes
```

* Cleaning column SCRAP

```python
data[data['SCRAP'].isnull()]
```

```python
data['SCRAP'].isnull().value_counts()
```

```python
data['SCRAP'] = pd.to_numeric(data['SCRAP'], errors='coerce')
```

```python
data['SCRAP'].isnull().value_counts()
```

```python
data['SCRAP'] = data['SCRAP'].astype('float')
data.dtypes
```
